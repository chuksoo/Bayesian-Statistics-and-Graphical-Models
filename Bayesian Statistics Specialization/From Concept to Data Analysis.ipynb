### Week 1 - Probability

There are three different framework for defining probabilities.
- Classical framework: works well when we have equally likely outcome
- Frequentist framework: requires us to have a hypothetical sequence of event. This is the probability derived from relative frequency perspective.
- Bayesian probability: requires personal perspective. This probability is based on a synthesis of evidence and personal judgment regarding that evidence.

The expected value of a random variable X is a weighted average of values X can take, with weights given by the probabilities of those values. If X can take on only a finite number of values (say, x1, x2, ... xn), we can calculate the expected value as 
\begin{equation}
E(X) = \sum_{i=1}^{n} x_i . P(X = x_i)
\end{equation}

### Bayes' theorem

Conditional probability is given by:
\begin{equation}
P(A|B) = \frac{P(A \cap B)}{P(B)} 
\end{equation}
This is denoted as the probability of A given that event B happened.

Given the table:

|  | Female | Not Female | Total |
| ---- | ---- | ---- | ---- |
| CS Major | 4 | a | 12 |
| Not CS major | b | c | d|
| Total | 9 | e | 30 |

```python
total_F = 9
total_CS = 12
total_F_n_CS = 4
total_population = 30
proba_F = total_F / total_population 
print('Probability of female is {:.2}'.format(proba_F))

proba_CS = total_CS / total_population
print('Probability of CS major is {:.2f}'.format(proba_CS))
```

```python
proba_F_n_CS = total_F_n_CS / total_population
print('Probability of female and CS major is {:.2f}'.format(proba_F_n_CS))
```

#### Conditional probability question

### Week 2 - Frequentist, Bayesian Inference, Likelihood and Maximum likelihood functions

The likelihood function is given by:
\begin{equation}
L(\theta \mid y) = \prod_{i=1}^{n} \theta^{y_i}(1-\theta)^{1-y_i}
\end{equation}

Plotting the log-likelihood for Bernoulli

$ \mathcal{L}(\theta) = \sum_i^n y_i ln(\theta) + \sum_i^n (1 - y_i)ln(1 - \theta) $

$ \mathcal{L}(\theta) = y ln(\theta) + (n - y)ln(1 - \theta) $

### Problem Statement

A hospital with 400 heart attack patients, of whom died within 30 days, and 328 are still alive.

Let's plot the likelihood function. Using a Bernoulli likelihood. The likelihood is a function of the mortality rate $\theta$

$ Y \sim B(\theta) $

$ f(x | \theta) = \theta^x (1 - \theta)^{1 - \theta} $

$ \mathcal{L}(\theta | y) = \prod_{i=1}^{n} \theta^{y_i} (1 - \theta)^{1 - y_i} $

$ \mathcal{L}(\theta | y) = \theta^{y} (1 - \theta)^{n - y} $

where: 

$y = \text{mortalities}$ 

$n = \text{Heart attack patients}$

```python
import os
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline

# the likelihood function in Python
def likelihood(n, y, Θ):
    """
    this function returns the 
    likelihood function
    """
    estimate_likelihood = Θ**y * (1 - Θ)**(n-y)
    return estimate_likelihood

# the log-likelihood
def log_likelihood(n, y, Θ):
    """
    this defines the log
    likelihood function
    """
    log_like = y * np.log(Θ) + (n - y) * np.log(1 - Θ)
    return log_like


# compute the likelihood function
Θ = np.arange(0.01, 0.99, 0.01)
estimate_likelihood = likelihood(400, 72, Θ)
theta_hat = np.max(estimate_likelihood)

# plot the likelihood function
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(Θ, estimate_likelihood, 
        alpha=0.5,
        marker='o',
        markersize=8)
ax.grid()
ax.set_xlabel('Θ', fontsize=14)
ax.set_ylabel('$L(Θ|y)$', fontsize=14)
ax.axis(xmin=0, ymin=0)
plt.axvline(x = 0.18, color = 'red', linestyle='--', lw=1.5, ymax = max(Θ))
plt.title('Plot of likelihood function')
plt.show();
print('The maximum likelihood obtained is {}'.format(theta_hat))
```

```python
# log-likelihood
log_like = log_likelihood(400, 72, Θ)
log_theta_hat = np.max(log_like)

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(Θ, log_like, 
        alpha=0.5,
        marker='o',
        markersize=8)
ax.grid()
ax.set_xlabel('Θ', fontsize=14)
ax.set_ylabel('$l(Θ)$', fontsize=14)
ax.axis(xmin=Θ.min(), ymin=log_like.min())
plt.axvline(x = 0.18, color = 'red', linestyle='--', lw=1.5, ymax = max(Θ))
plt.title('Plot of log-likelihood function')
plt.show();
print('The maximum log-likelihood obtained is {:.2f}'.format(log_theta_hat))
```

The equatopm to compute the Residual standard error is

$ RSE = \sqrt{\frac{1}{n - 2}RSS} $

Where:

$ RSS = \sum_{i = 1}^n(y - \hat{y}) $

```python
# Computing the residual standard error
RSS = np.sum(np.square(o_ring - o_ring_pred))
rse = np.sqrt(RSS / (o_ring.shape[0] - 2))
print('The residual standard error of the fit is: {}'.format(rse))
```

**The Challenger launch was at 31 degrees Fahrenheit how much o-ring damage would we predict?**

```python
y_hat = reg.predict(np.array([31]).reshape(-1,1))
print("The predicted value for damage is: {}".format(y_hat))
```

```python
import statsmodels.api as sm

N = temp.shape[0]
p = temp.shape[1] + 1  # Plus one because LinearRegression adds an intercept term

X_with_intercept = np.empty(shape=(N, p), dtype=float)
X_with_intercept[:, 0] = 1
X_with_intercept[:, 1:p] = temp

ols = sm.OLS(o_ring, X_with_intercept)
ols_result = ols.fit()
ols_result.summary()
```

The fitted intercept is 18.365 and the fitted slope is minus 0.24. The slope gives the relationship between temperature and predicted damage. It's saying approximately, for every four degree decrease in temperature, we expect to see a one unit increase in damage. That makes sense with this negative slope.

```python
# Computing residuals
residuals = o_ring - o_ring_pred
# Sum of squares
residual_sum_of_squares = residuals.T @ residuals
sigma_squared_hat = residual_sum_of_squares[0, 0] / (N - p)
var_beta_hat = np.linalg.inv(X_with_intercept.T @ X_with_intercept) * sigma_squared_hat
for p_ in range(p):
    standard_error = var_beta_hat[p_, p_] ** 0.5
    print(f"SE(beta_hat[{p_}]): {standard_error}")
# Note that these values coincide with the ones reported above!
```

95% posterior interval for the slope:

There are 21 degrees of freedom in the residuals and the residual standard error is 0.06349.

```python
from scipy.stats import t

df = 21
rv = t(df)
# Intercept + std error param * quantile t student 
v1 = -0.24337 - 0.06349*rv.ppf(.975)
v2= -0.24337 + 0.06349*rv.ppf(.975)
print("The lower bound is: {}".format(v1))
print("The upper bound is: {}".format(v2))
```

This interval does not include 0 so we can be pretty sure that there is a negative relationship between temperature and damage. Note that these intervals are the same as the frequent confidence intervals when we're using the standard reference prior for Bayesian analysis.

We can also think about a prediction interval. A posterior prediction interval. So if we are going to launch at 31 degrees, what is a 95% predictive interval for what we might see in damage? This is going to be a larger interval than just a fitted interval.

Recall that the equation for the scale or variance of a new prediction is:

$SE_r \sqrt{1 + \frac{1}{n} + \frac{(n - 1)(x^* - \bar{x})}{S_x^2}} $

$SE_r = \text{Residual standard error of regression} $

$S_x^2 = \text{Sample variance} $

```python
# The mean value is the one shown above, for the lower interval we have:
# T student distribution
df = 21
rv = t(df)

n = 23 # Number os samples temp.shape[0]
x_star = 31 # Value to predict
mean_temp = np.mean(temp) # Mean Temp
sample_variance = np.var(temp) # Variance Temp
lower_bound = 10.82052 - ((rse * rv.ppf(.975)) * np.sqrt(1 + (1/n) + (np.power(x_star - mean_temp, 2) / (n - 1) / sample_variance)))
upper_bound = 10.82052 + ((rse * rv.ppf(.975)) * np.sqrt(1 + (1/n) + (np.power(x_star - mean_temp, 2) / (n - 1) / sample_variance)))
print("The lower bound for the predictive interval is: {}".format(lower_bound))
print("The upper bound for the predictive interval is: {}".format(upper_bound))
```

A 95% predictive interval goes from about 4.05 to 17.59.

Posterior probability that damage index is greater than zero at 31F? As the posterior distribution works as a T student.

So it's a $t$ distribution I'm going to ask what's probably a T with this center and this scale is bigger than 0. The rv.cdf command will give us the probability it's less than 0 and so we take 1 minus that to get the probability it's bigger than 0. Probably, that the damage will be bigger than 0 is 0.998, very close to one. So it’s extremely likely that there will be damage if we are to launch at 31 degrees Fahrenheit.

```python
prob = 1 - rv.cdf((0 - 10.82052) / (rse * np.sqrt(1+ (1/n) +(np.power(x_star - mean_temp, 2)/ (n - 1) / sample_variance))))
print("The posterior probability that the damage index is greather than zero is: {}".format(prob))
```
